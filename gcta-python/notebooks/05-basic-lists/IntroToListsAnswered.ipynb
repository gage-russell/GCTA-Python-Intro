# Lists

Lists are used to store multiple items in a single variable.

Lists are one of 4 built-in data types in Python used to store collections of data, the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.

Lists are created using square brackets

```python
# Here is an example of a list

my_list = [1,2,3,4]
print(my_list)
```

```python
# Create a list containing the values 1,2,3 and check its type using the type function

my_list = [1,2,3]
type(my_list)
```

```python
# Create a list containing the values 1,2,3 and check that it is a list using the isinstance function

my_list = [1,2,3]
isinstance(my_list, list)
```

# Indexing Lists

Lists are ordered objects. Ordered meaning that each item has a defined order.

This means that we can select specific items from a list using its index.

Indexes start at 0 in python. This means that the first item in a list will be of index 0.

Lists are indexed using square brackets. to get the first item in the list, my_list, you could do `my_list[0]`

Multiple items from a list can be gotten using slicing. Slicing follows the same syntax as indexing except that you provide a range of indices [start `:` stop].

Slicing selects the start as the first index and goes up to but not including the index corrsponding to stop.

The first 3 items in my_list can be found by: `my_list[0:3]`
This could also be written: `my_list[:3]`
not specifying a value on either side of the colon `:` means that it will select every item up to the other side of the expression.

We can also backwards index lists. This means that if we index with negative numbers, we are selecting items from right to left.

`my_list[-1]` will return the last item in a list

```python
# Create a list containing values 1,2,3 and then print the first item in the list

my_list = [1,2,3]
print(my_list[0])
```

```python
# Create a list containing values 1,2,3 and then print the first two items in the list

my_list = [1,2,3]
print(my_list[0:2])
```

```python
# Create a list containing values 1,2,3,4 and then print the the two middle numbers in the list

my_list = [1,2,3,4]
print(my_list[1:3])
```

```python
# Create a list containing values 1,2,3,4 and then print the last item

my_list = [1,2,3,4]
print(my_list[-1])
```

```python
# Create a list containing values 1,2,3,4 and then print the last two items

my_list = [1,2,3,4]
print(my_list[-2:])
```

# List Mutability

Lists are mutable. Meaning they are changeable.

You can change, add, and remove items in a list after it has been created.

Specifically, the del command can be used to delete items in a list. Lists also have built in methods for adding and removing items.

```python
# Create a list containing values 1,2,3,4 and then change the value of of the second item in the list (2) to be 9. 
# Use the indexing we learned above and the assignment operator (=) to do this

my_list = [1,2,3,4]
my_list[1] = 9
print(my_list)
```

# List Length

The python built-in function, `len` can be used to find the length of a list: `len(my_list)`

```python
# create a list containing values 1,2,3,4 then print the length of it using the len function

my_list = [1,2,3,4]
print(len(my_list))
```

# Adding Lists

Lists are operated on very similar to strings.

Including summation. When two lists are added together, they will be concatenated

```python
# create two lists. Add them together. Print the result.

my_list = [1,2,3]
my_other_list = [4,6,2]

result = my_list + my_other_list
print(result)
```

# List Methods

In Python, a method is a function that is available for a given object because of the object's type.

Methods are called using .notation -- `object.method_name()`

List methods can be found here: https://www.w3schools.com/python/python_ref_list.asp



```python
# Create a list containing the values 1,2,3 then use the append method to add 4 to the list

my_list = [1,2,3]
my_list.append(4)
print(my_list)
```

```python
# create a list containing the values 1,2,3 then use the clear method to remove all of its values

my_list = [1,2,3]
my_list.clear()
print(my_list)
```

```python
# create a list containing the values 1,1,1,3 then use the count method to see how many items have the value 1

my_list = [1,1,1,3]
print(my_list.count(1))
```

```python
# create a list containing the values 5,9,8,5 then use the sort method to sort it ascendingly

my_list = [5,9,8,5]
my_list.sort()
print(my_list)
```

```python
# create a list containing the values 'cat','dog' then use the reverse method to reverse its order

my_list = ['cat','dog']
my_list.reverse()
print(my_list)
```
