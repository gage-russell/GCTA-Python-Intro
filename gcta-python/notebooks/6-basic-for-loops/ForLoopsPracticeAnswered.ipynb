```python
my_list = [1,2,3]

my_new_list = []

for num in my_list:
    
    num_sqd = num**2
    
    my_new_list.append(num_sqd)
    
print(my_new_list)
```

```python
my_list = ['hank', 'sally', 'phil']

my_new_list = []

for name in my_list:
    
    my_letter = name[0]
    
    my_new_list.append(my_letter)
    
print(my_new_list)
```

```python
my_list = ['hank', 'sally', 'phil']

my_new_list = []

i = 0

for name in my_list:
    
    my_letter = name[i]
    
    my_new_list.append(my_letter)
    
    i += 1
    
print(my_new_list)
```

```python
"""
Q: How many iterations will there be?

Q: What variables do I need to track as I iterate? -- what variables are changing within the loop?

ITERATION 1:
    Variables:
        
        name: 'hank'
        
        i: 0
        
        my_letter: 'hank'[0] = 'h'
        
        my_new_list: [] -> ['h']
        
ITERATION 2:
    Variables:
        
        name: 'sally'
        
        i: 1
        
        my_letter: 'sally'[1] = 'a'
        
        my_new_list: ['h'] -> ['h', 'a']
        
ITERATION 3:
    Variables:
        
        name: 'phil'
        
        i: 2
        
        my_letter: 'phil'[2] = 'i'
        
        my_new_list: ['h', 'a'] -> ['h', 'a', 'i']
        
----------------
printing ['h', 'a', 'i']
"""
```

```python
my_list = [5,10,15]
my_other_list = ['a', 'b', 'c']

my_dictionary = {}

iter_count = len(my_list)

for i in range(iter_count):
    
    my_dictionary[my_other_list[i]] = my_list[i]
    
print(my_dictionary)
    
```
