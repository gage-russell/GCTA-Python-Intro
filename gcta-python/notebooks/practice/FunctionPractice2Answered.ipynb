```python
my_dict = {
    'a': [2, 4],
    'b': [6, 8]
}

def my_function(d):
    num = 0
    
    for i in d.values():
        num += sum(i)
        
    return num

foo = my_function(my_dict)

# create a variable foo such that this cell prints True

z = 20

print(foo == z)
```
