```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = 'https://www.indeed.com/jobs?q=python&l=Austin%2C%20TX'
#conducting a request of the stated URL above:
page = requests.get(URL)
#specifying a desired format of “page” using the html parser - this allows python to read the various components of the page, rather than treating it as one long string.
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# put this into a file so that we can explore and figure out where the job listings are and parse them out
with open("soup_html_example_output.html", "w", encoding="utf-8") as file:
    file.write(str(soup))
```

```python
# now that we know whats inside of a job listing, we can start parsing
mydivs = soup.find_all("table", {'class':'jobCard_mainContent', 'role':'presentation'})
for result in mydivs:
    for listing in result.find_all('tr'):
        # Get the Job Title
        title_heading = listing.find('h2', {'class':['jobTitle jobTitle-color-purple', 'jobTitle jobTitle-color-purple jobTitle-newJob']})
        if title_heading:
            title_span = title_heading.find_all('span')
            if title_span:
                if len(title_span) > 0:
                    title_span = title_span[-1]
                    title = title_span.contents[0]
                    print(title)
```

```python
# now that we know whats inside of a job listing, we can start parsing
mydivs = soup.find_all("table", {'class':'jobCard_mainContent', 'role':'presentation'})
for result in mydivs:
    for listing in result.find_all('tr'):
        # Get the Job Title
        title_heading = listing.find('h2', {'class':['jobTitle jobTitle-color-purple', 'jobTitle jobTitle-color-purple jobTitle-newJob']})
        if title_heading:
            title_span = title_heading.find_all('span')
            if title_span:
                if len(title_span) > 0:
                    title_span = title_span[-1]
                    title = title_span.contents[0]
                    print(title)
        
        # Get the company
        company_a = listing.find('a', {'data-tn-element': 'companyName'})
        if company_a:
            if len(company_a) > 0:
                company_name = company_a.contents[0]
                print(company_name)
        else:
            company_a = listing.find('span', {'class':'companyName'})
            if len(company_a) > 0:
                company_name = company_a.contents[0]
                print(company_name)
        
        # Get the company rating
        company_rating_span = listing.find('span', {'class':'ratingNumber'})
        if company_rating_span:
            company_rating_span_span = company_rating_span.find('span')
            if len(company_rating_span_span) > 0:
                company_rating = company_rating_span_span.contents[0]
                print(company_rating)
                
        # Get the company location
        company_location_div = listing.find('div', {'class':'companyLocation'})
        if company_location_div:
            if len(company_location_div) > 0:
                company_location = company_location_div.contents[0]
                print(company_location)
```

```python
# thats great, but lets do something beyond printing the results

jobs = []

mydivs = soup.find_all("table", {'class':'jobCard_mainContent', 'role':'presentation'})
for result in mydivs:
    for listing in result.find_all('tr'):
        # print(listing.prettify())
        # Get the Job Title
        title_heading = listing.find('h2', {'class':['jobTitle jobTitle-color-purple', 'jobTitle jobTitle-color-purple jobTitle-newJob']})
        if title_heading:
            title_span = title_heading.find_all('span')
            if title_span:
                if len(title_span) > 0:
                    title_span = title_span[-1]
                    title = title_span.contents[0]
        
        # Get the company
        company_a = listing.find('a', {'data-tn-element': 'companyName'})
        if company_a:
            if len(company_a) > 0:
                company_name = company_a.contents[0]
        else:
            company_a = listing.find('span', {'class':'companyName'})
            if len(company_a) > 0:
                company_name = company_a.contents[0]
        
        # Get the company rating
        company_rating_span = listing.find('span', {'class':'ratingNumber'})
        if company_rating_span:
            company_rating_span_span = company_rating_span.find('span')
            if len(company_rating_span_span) > 0:
                company_rating = company_rating_span_span.contents[0]
                
        # Get the company location
        company_location_div = listing.find('div', {'class':'companyLocation'})
        if company_location_div:
            if len(company_location_div) > 0:
                company_location = company_location_div.contents[0]
                
        listing_info = {'title': title, 'company': company_name, 'company_rating': company_rating, 'location': company_location}
        jobs.append(listing_info)
        
        # so that we don't mix up results
        title = None
        company_name = None
        company_rating = None
        company_location = None
```

```python
# Cool! lets check out our new list of jobs
jobs
```

```python
# how can we make this more useable?
job_df = pd.DataFrame(jobs)
```
